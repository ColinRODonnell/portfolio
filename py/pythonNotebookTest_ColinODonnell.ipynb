3b. Prepare a code block that depends on running the previous code block and then reads from each directory of files and calculates what's needed to indicate **lexical diversity**.
    Running the cell that you prepare should reveal the lexical diversity of each set of files. *(3 pts)*

```python
def lexical_diversity_file(filepath):
    with open(filepath, encoding="utf-8") as f:
        text = f.read()
    
    doc = nlp(text)
    #spaCy does this part 
    words = [token.text.lower() for token in doc if token.is_alpha]
    
    return len(set(words)) / len(words)

first = lexical_diversity_file("first-episode/ATE1.txt")
final = lexical_diversity_file("final-episode/ATE26.txt")

print("First episode lexical diversity:", first)
print("Final episode lexical diversity:", final)
```
